```python
# Import files
import pandas as pd
import plotly.express as px

happiness_index = pd.read_csv("WHI_2022.csv")

# Remove special characters from the country names
happiness_index["Country"] = happiness_index["Country"].str.replace("*", "")

happiness_index["Country"].unique()


# Data about sector distribution
## https://www.cia.gov/the-world-factbook/about/archives/2023/field/gdp-composition-by-sector-of-origin/
## https://wdi.worldbank.org/table/4.2

happiness_index.head()
```

```python
happiness_index[['RANK', 'Country', 'Happiness score', 'Explained by: GDP per capita',
       'Explained by: Social support', 'Explained by: Healthy life expectancy',
       'Explained by: Freedom to make life choices',
       'Explained by: Generosity', 'Explained by: Perceptions of corruption']].head()
```

```python
country_data = pd.read_csv("country_data.csv").drop_duplicates().reset_index()

# Make country names the same
country_data["Country"] = country_data["Country"].str.replace("*", "")
country_data["Country"] = country_data["Country"].replace({"Taiwan (Province of China)": "Taiwan Province of China",
                                                           "Hong Kong (SAR)": "Hong Kong S.A.R. of China",
                                                           "Congo (DRC)": "Congo"})


country_data["Post-War Alignment"] = country_data["Post-War Alignment"].replace({"Soviet Bloc (USSR)": "Soviet Bloc",
                                                                                "Neutral/Non-Aligned": "Neutral",
                                                                                "Western Bloc (NATO)": "Western Bloc",
                                                                                "Soviet Bloc (USSR-aligned)": "Soviet Bloc",
                                                                                "Western Bloc (France-aligned)": "Western Bloc",
                                                                                "Western Bloc (U.K.-aligned)": "Western Bloc",
                                                                                "Western Bloc (U.S.-aligned)": "Western Bloc",
                                                                                "Non-Aligned Movement": "Neutral",
                                                                                "Neutral (Post-Yugoslav)": "Neutral",
                                                                                "Western Bloc (U.S.-aligned, Turkey-aligned)": "Western Bloc",
                                                                                "Soviet Bloc (Warsaw Pact)": "Soviet Bloc",
                                                                                "Neutral": "Neutral",
                                                                                "Western Bloc (NATO, West Germany)": "Western Bloc",
                                                                                "Soviet Bloc (Warsaw Pact, former Czechoslovakia)": "Soviet Bloc",
                                                                                "Non-Aligned (former Yugoslavia)": "Neutral",
                                                                                "Western Bloc (Joined NATO in 1982)": "Western Bloc"})


country_data["Majority Religion"] = country_data["Majority Religion"].replace({"Christianity (Catholic/Protestant)": "Christianity",
                                                                                "Christianity (Protestant/Catholic)": "Christianity",
                                                                                "Islam (Sunni)": "Islam",
                                                                                "Buddhism/Folk Religion": "Buddhism",
                                                                                "Buddhism/Christianity": "Buddhism",
                                                                                "Shinto/Buddhism": "Buddhism",
                                                                                "Christianity/Buddhism": "Christianity",
                                                                                "Islam/Christianity (Orthodox)": "Islam",
                                                                                "Islam/Christianity": "Islam",
                                                                                "Islam (Shia)": "Islam",
                                                                                "Christianity (Orthodox)": "Christianity",
                                                                                "Christianity (Catholic)": "Christianity",
                                                                                "Christianity (Protestant)": "Christianity",
                                                                                "Christianity (Lutheran)": "Christianity",
                                                                                "Christianity (Anglican)": "Christianity",
                                                                                "Christianity/Indigenous": "Christianity",
                                                                                "Christianity/Islam": "Christianity",
                                                                                "Folk Religion/Buddhism": "Buddhism"})
```

```python
# Merge both databases
merged_df = pd.merge(happiness_index, country_data, on="Country", how="left").dropna()      # Drop NA to remove the last row of happiness_index
merged_df.head()
```

```python
income_levels = pd.read_csv("income_levels.csv")[["TableName", "IncomeGroup"]]
income_levels.rename(columns={'TableName': 'Country', "IncomeGroup": "Income_group"}, inplace=True)

income_levels["Country"] = income_levels["Country"].replace({"Slovak Republic": "Slovakia",
                                                             "Korea, Rep.": "South Korea",
                                                             "Kyrgyz Republic": "Kyrgyzstan",
                                                             "Viet Nam": "Vietnam",
                                                             "North Cyprus": "Cyprus",
                                                             "Russian Federation": "Russia",
                                                             "Hong Kong SAR, China": "Hong Kong S.A.R. of China",
                                                             "Gambia, The": "Gambia",
                                                             "Lao PDR": "Laos",
                                                             "Congo, Dem. Rep": "Congo",
                                                             "Venezuela, RB": "Venezuela",
                                                             "Iran, Islamic Rep.": "Iran",
                                                             "Eswatini": "Eswatini, Kingdom of",
                                                             "Egypt, Arab Rep.": "Egypt",
                                                             "Yemen, Rep.": "Yemen",
                                                             "Côte d'Ivoire": "Ivory Coast",
                                                             "Türkiye": "Turkey",
                                                             "Cyprus": "North Cyprus"})


# Merge both databases
merged_df_2 = pd.merge(merged_df, income_levels, on="Country", how="left")

merged_df_2.head()

```

```python
# Cast the correct datatype on these columns (float i/o str)
columns_to_change= ["Happiness score",
                    "Whisker-high",
                    "Whisker-low",
                    "Dystopia (1.83) + residual",
                    "Explained by: GDP per capita",
                    "Explained by: Social support",
                    "Explained by: Healthy life expectancy",
                    "Explained by: Freedom to make life choices",
                    "Explained by: Generosity",
                    "Explained by: Perceptions of corruption"]

for column in columns_to_change:
    merged_df_2[column] = merged_df_2[column].str.replace(',', '.', regex=False)  # Replace , with .
    merged_df_2[column] = pd.to_numeric(merged_df_2[column])  # Convert to numeric type (float)


merged_df_2 = merged_df_2[merged_df_2["Country"]!="Israel"] # Remove Israel as it doesn't provide general knowledge

```

```python
# Importance of Perceived Freedom by Political System
px.box(merged_df_2,
             x='Political System',
             y='Explained by: Freedom to make life choices',
             color="Political System",
             points="all",
             hover_data="Country",
             title="Importance of Perceived Freedom by Political System").show()

# Does this mean that freedom is more important where they don't have as much as democracy
```

```python
merged_df_2["total_happiness"] = merged_df_2["Dystopia (1.83) + residual"] + merged_df_2["Explained by: GDP per capita"] + merged_df_2["Explained by: Social support"] + merged_df_2["Explained by: Healthy life expectancy"] + merged_df_2["Explained by: Freedom to make life choices"] + merged_df_2["Explained by: Generosity"] + merged_df_2["Explained by: Perceptions of corruption"]

merged_df_2["normalized_GDPPC"] = merged_df_2["Explained by: GDP per capita"]/merged_df_2["total_happiness"]
merged_df_2["normalized_Social_Support"] = merged_df_2["Explained by: Social support"]/merged_df_2["total_happiness"]
merged_df_2["normalized_Healthy_life_expectancy"] = merged_df_2["Explained by: Healthy life expectancy"]/merged_df_2["total_happiness"]
merged_df_2["normalized_Freedom"] = merged_df_2["Explained by: Freedom to make life choices"]/merged_df_2["total_happiness"]
merged_df_2["normalized_Generosity"] = merged_df_2["Explained by: Generosity"]/merged_df_2["total_happiness"]
merged_df_2["normalized_Corruption"] = merged_df_2["Explained by: Perceptions of corruption"]/merged_df_2["total_happiness"]
```

```python
# Importance of Perceived Freedom by Political System
px.box(merged_df_2,
             x='Political System',
             y='normalized_Freedom',
             color="Political System",
             points="all",
             hover_data="Country",
             title="Importance of Perceived Freedom by Political System").show()
```
